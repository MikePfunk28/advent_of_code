## Validate UUID


```python
# at least 2 Uppercase English alphabet chars
# 3 digits
# alphanumeric only, A-Za-z0-9
# no repeat chars
# length 10

"""
if index in list:
    isupper()
    count+=1

    isalnum()


"""
if __name__ == '__main__':

    s = int(input())
    for _ in range(s):
        s = input().strip()
        upper_count = sum(1 for c in s if c.isupper())
        digit_count = sum(1 for c in s if c.isdigit())
        alphanumeric_only = all(c.isalnum() for c in s)
        no_repeats = len(set(s)) == len(s)
        length_ok = len(s) == 10

        if upper_count >= 2 and digit_count >= 3 and alphanumeric_only and no_repeats and length_ok:
            print("Valid")
        else:
            print("Invalid")
```

UUID validation, essentially using a set and comparing it to the original list, will check for dupes.  isalnum will make sure its alphanumeric, upper_count, and digit_count both count the upper, min 3, and the digits and validates the length is 10.
