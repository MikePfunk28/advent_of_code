```python
# 1: Two Sum
# return indices of the two numbers such that they add up to a specific target
# 2, 3, 4, 5, 6, 7, 8, 9, 10
class Solution:
    def __init__(self):
        self.nums = [2, 3, 4, 5, 6, 7, 8, 9, 10]
        self.target = None


    def twoSum(self, target):
        self.target = target
        for i in range(len(self.nums)):
            for j in range(i + 1, len(self.nums)):
                if self.nums[i] + self.nums[j] == self.target:
                    return [i, j]
        return None

    def multipleTwoSum(self, target):
        self.target = target
        result = []
        for i in range(len(self.nums)):
            for j in range( i, len(self.nums)):
                if self.nums[i] + self.nums[j] == self.target:
                    result.append([i, j])
        return result

    solution = Solution()
    print(solution.twoSum(10))
    print(solution.multipleTwoSum(10))


```

```python
import re

if __name__ == "__main__":
# Read N (the number of inputs)
    n = int(input())

for _ in range(n):
    phone_number = input()
    # re.match checks if the pattern matches from the beginning
    # We use .bool() logic or just 'if' to see if it found a match
    if re.match(r'^[789]\d{9}$', phone_number):
        print("YES")
    else:
        print("NO")
```
