````python
class Solution:
    def romanToInt(self, s: str) -> int:
        values = {
            "I": 1,
            "V": 5,
            "X": 10,
            "L": 50,
            "C": 100,
            "D": 500,
            "M": 1000,
        }

        total = 0
        previous = 0

        for char in reversed(s):
            current = values[char]

            if current < previous:
                total -= current
            else:
                total += current

            previous = current

        return total
    
if __name__ == "__main__":
    solution = Solution()
    s = input().split()
    print(solution.romanToInt(s[0]))
    print(solution.romanToInt("III"))  # Output: 3
    print(solution.romanToInt("LVIII"))  # Output: 58
    print(solution.romanToInt("MCMXCIV"))  # Output: 1994
````

**input()** grabs the text: It reads the very first line of the test case as a raw text string: "2 2".

**.split()** chops it up: It splits that string by the space in the middle, turning it into a list of two separate text strings: ['2', '2'].

**map()** does the conversion: Remember that factory assembly line analogy? map takes the int function and applies it to every item in that list. It turns the string '2' into the math number 2, and the second string '2' into the math number 2.

**n, m** =  unpacks the results: Because map spit out exactly two integers, Python automatically assigns the first integer to the variable n and the second integer to the variable m.

## The Three Parts of the Factory
The Conveyor Belt (The Iterable): This is your list of raw data. In our HackerRank problem, the raw data is the list of text strings we got from splitting the input: ['2', '2'].

The Tool (The Function): This is the specific machine or action you want to apply to your materials. For our code, the tool is the int function, which stamps a piece of text and turns it into a math number.

The Supervisor (map): This is the map() function itself. The supervisor doesn't do the stamping manually. Instead, you hand the supervisor the Tool and point them to the Conveyor Belt.

## How the Assembly Line Runs
When you write the code map(int, ['2', '2']), you are giving an order to the supervisor:

"Hey map, take this int tool, stand next to the conveyor belt, and use the tool on every single item that rolls by."

The supervisor then turns on the belt:

The first item '2' (text) rolls down. The supervisor uses the int tool on it. Out pops 2 (number).

The second item '2' (text) rolls down. The supervisor uses the int tool on it. Out pops 2 (number).

**map** replaces the for loop here.
