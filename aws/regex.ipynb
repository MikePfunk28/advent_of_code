```python
def is_valid_hex_color(text):
    if not text:
        return False
        
    # Strip optional CSS hash or programming prefix
    if not text.startswith("#"):
        return False
        
    # Skip the leading '#' for validation    
    hex_data = text[1:]
        
    # Strictly enforce 3 or 6 character lengths
    if len(hex_data) not in (3, 6):
        return False
        
    # Verify all remaining characters are hex digits
    valid_hex = "0123456789abcdefABCDEF"
    return all(char in valid_hex for char in hex_data)



if __name__ == "__main__":

    # Test cases
    print(is_valid_hex_color("#FFF"))     # True (Length 3)
    print(is_valid_hex_color("1a2b3c"))   # True (Length 6)
    print(is_valid_hex_color("#1234"))    # False (Length 4 is invalid)
    print(is_valid_hex_color("12G"))      # False ('G' is not hex)
```

```python
import sys
import io

input_stdin = """
11
#BED
{
    color: #FfFdF8; background-color:#aef;
    font-size: 123px;
    background: -webkit-linear-gradient(top, #f9f9f9, #fff);
}
#Cab
{
    background-color: #ABC;
    border: 2px dashed #fff;
}  
"""

if input_stdin:
    sys.stdin = io.StringIO(input_stdin.strip())

if __name__ == '__main__':
    # Step 1: Read all lines from HackerRank's input
    lines = sys.stdin.read().splitlines()
    
    # Start with our switch turned OFF
    inside_block = False
    
    # Step 2: Loop through every line
    for line in lines:
        if '{' in line:
            inside_block = True   # Switch ON
            continue
        if '}' in line:
            inside_block = False  # Switch OFF
            continue
            
        # Only look for colors if we are inside a CSS property block
        if inside_block:
            # Step 3: Search the line character-by-character for '#'
            for i in range(len(line)):
                if line[i] == '#':
                    
                    # Step 4: Gather characters right after the '#'
                    potential_hex = ""
                    count = 0
                    
                    for j in range(i + 1, len(line)):
                        char = line[j]
                        
                        # Check if it's a valid hex digit
                        if char.isdigit() or ('a' <= char <= 'f') or ('A' <= char <= 'F'):
                            potential_hex += char
                            count += 1
                        else:
                            # Hit a space, semicolon, or junk -> stop gathering
                            break
                    
                    # Step 5: Check if the count is exactly 3 or 6
                    if count == 3 or count == 6:
                        print("#" + potential_hex)
```
