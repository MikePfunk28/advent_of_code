## Parsing methods

**.handle_start(tag, attrs)**
This method is called to handle the start tag of the element.
(For example: <div class='marks'>)

The tag argument is the name of the tag converted to lowercase.

The attr argument is a list of (name, value) pairs containing the attributes found inside the tag's <> brackets.

**.handle_endtag(tag)**
This method is called to handle the end tag of an element.
(For example: </div>)
The tag argument is the name of the tag converted to lowercase.

**.handle_startendtag(tag.attr)**
This method is called to handle an empty tag of an element.
(For example: <br />)
The tag element is the name of the tag converted to lowercase.

The attr element is a list of (name, value) pairs containing the  attributes found inside the tag's <> brackets.

## Inheritance

```python
class Parent:
    def __init__(self):
        self.ready = True

class Child(Parent):
    def __init__(self):
        super().__init__()
        self.name = "child"
```

### Inheritance - 'is a' relationship

```python
class MyHTMLParser(HTMLParser):
    pass
```

### Association - 'Uses/Knows about' relationship

```python
class Student:
    def __init__(self, teacher):
        self.teacher = teacher
```

### Aggregation - weaker 'has a' relationship, where contained object can exist independently.

```python
class Department:
    def __init__(self, professor):
        self.professor = professor
```
A Department has a Professor, but the Professor can exist without that department.

### Composition - A stronger 'has a' relationship where the contained object is owned by the container.

```python
class House:
    def __init__(self):
        self.room = Room()
```

A House has a Room, and the Room is created as part of the House.

Inheritance  -> is a
Association  -> uses / knows a
Aggregation  -> has a, but independent
Composition  -> has a, owned lifecycle
