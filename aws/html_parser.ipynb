## Parsing methods

**.handle_start(tag, attrs)**
This method is called to handle the start tag of the element.
(For example: <div class='marks'>)

The tag argument is the name of the tag converted to lowercase.

The attr argument is a list of (name, value) pairs containing the attributes found inside the tag's <> brackets.

**.handle_endtag(tag)**
This method is called to handle the end tag of an element.
(For example: </div>)
The tag argument is the name of the tag converted to lowercase.

**.handle_startendtag(tag.attr)**
This method is called to handle an empty tag of an element.
(For example: <br />)
The tag element is the name of the tag converted to lowercase.

The attr element is a list of (name, value) pairs containing the  attributes found inside the tag's <> brackets.

## Parse out the comment and label it multi line or single line, following the format specified.

````python
import sys
from html.parser import HTMLParser

class MyHTMLParser(HTMLParser):
    def __init__(self):
        super().__init__()
        self.multiline_comment = []
        self.singleline_comment = []
    
    def handle_comment(self, data):
        if '\n' in data:
            self.multiline_comment.append(data)
            print(">>> Multi-line Comment")
        else:
            self.singleline_comment.append(data)
            print(">>> Single-line Comment")
        print(data)
        
if __name__ == '__main__':
    n = int(input())
    # read it all at once, since the parser will handle it
    
    html_content = sys.stdin.read() 
    parser = MyHTMLParser()
    parser.feed(html_content)
````
