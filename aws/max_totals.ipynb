## map() vs starmap()
its similar to how function(a, b) and function(*c) works.

* - python unpacking operator

Standard **map()** expects you to pass in separate iterables for every arguement.  It then pulls one item from each iterable simulataneously to feed the function.
*   Data structure: Separate lists/iterables.
*   How it executes: pow(2, 5), then pow(3, 2)

**starmap()** Already pre-zipped or grouped into single iterable of tuples, takes a tuple and uses * operator to unpack the tuple into function arguments
*   Data structure: A single list of tuples (or other iterables).
*   How it executes: pow(*(2, 5)), then pow(*(3, 2))

## groupby() - How groupby Already Resets Everything
When you write for k, g in groupby(data):, Python takes your string '1222311' and chops it into completely separate, fresh lists every time the character changes.

If you just look at list(g) inside the loop, here is exactly what it contains on each pass:

Pass 1: ['1'] (It isolated the first '1')

Pass 2: ['2', '2', '2'] (It isolated the '2's)

Pass 3: ['3'] (It isolated the '3')

Pass 4: ['1', '1'] (It isolated the final '1's into a brand-new, reset list)

Notice how Pass 4 is just ['1', '1']? It doesn't remember the first '1' from Pass 1 at all. It is a completely fresh block.

The Built-in Solution: len()
Because list(g) is already a perfectly isolated list of just the current consecutive characters, you don't need a for loop to count them. You don't need sum() or if statements.

You just need to ask Python: "How many items are in this list?"

The function for that is len().

len(['2', '2', '2']) is 3.

len(['1', '1']) is 2.

````python
from itertools import groupby

data = input()

for k, g in groupby(data):
    # 1. Turn the group 'g' into a list so you can measure it
    group_list = list(g)
    
    # 2. Get the count simply by measuring the length of that list
    count = len(group_list)
    
    # 3. Print the count and the key 'k' in the format HackerRank wants
    # (Remember k is currently a string, so you might need to convert it to an int for the output!)
    print(f"({count}, {k})", end=' ')
````
