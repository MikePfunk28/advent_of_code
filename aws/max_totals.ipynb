## map() vs starmap()
its similar to how function(a, b) and function(*c) works.

* - python unpacking operator

Standard **map()** expects you to pass in separate iterables for every arguement.  It then pulls one item from each iterable simulataneously to feed the function.
*   Data structure: Separate lists/iterables.
*   How it executes: pow(2, 5), then pow(3, 2)

**starmap()** Already pre-zipped or grouped into single iterable of tuples, takes a tuple and uses * operator to unpack the tuple into function arguments
*   Data structure: A single list of tuples (or other iterables).
*   How it executes: pow(*(2, 5)), then pow(*(3, 2))


```python
bases = [2, 3, 10]
exponents = [5, 2, 3]

# map takes an item from 'bases' and an item from 'exponents'
results = list(map(pow, bases, exponents))
# Output: [32, 9, 1000]
```

```python
import itertools

paired_data = [(2, 5), (3, 2), (10, 3)]

# starmap takes one tuple at a time and unpacks it: pow(*tuple)
results = list(itertools.starmap(pow, paired_data))
# Output: [32, 9, 1000]
```
